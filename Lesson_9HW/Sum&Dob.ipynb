```python
from itertools import combinations_with_replacement

ALL_PAIRS = [(a, b) for a in range(2, 100) for b in range(a, 100)]

def factorizations(product):
    return [(a, product // a) for a in range(2, 100)
            if product % a == 0 and a <= product // a < 100]

def decompositions(s):
    return [(a, s - a) for a in range(2, s)
            if a <= s - a and s - a < 100 and a < 100]

def prod_doesnt_know(pair):
    a, b = pair
    return len(factorizations(a * b)) > 1

def sum_knew(pair):
    a, b = pair
    s = a + b
    return all(prod_doesnt_know(p) for p in decompositions(s))

def prod_now_knows(pair):
    a, b = pair
    p = a * b
    candidates = [q for q in factorizations(p) if sum_knew(q)]
    return len(candidates) == 1

def sum_now_knows(pair):
    a, b = pair
    s = a + b
    candidates = [q for q in decompositions(s) if sum_knew(q) and prod_now_knows(q)]
    return len(candidates) == 1

step1 = [p for p in ALL_PAIRS if prod_doesnt_know(p)]
step2 = [p for p in step1 if sum_knew(p)]
step3 = [p for p in step2 if prod_now_knows(p)]
step4 = [p for p in step3 if sum_now_knows(p)]

print(f"Step 1 (Prod не знає):              {len(step1)} пар")
print(f"Step 2 (Sum знав це):               {len(step2)} пар")
print(f"Step 3 (Prod тепер знає):           {len(step3)} пар")
print(f"Step 4 (Sum тепер теж знає):        {len(step4)} пар")

print("\nВідповідь")
for a, b in step4:
    print(f"  a = {a}, b = {b}   (сума = {a+b}, добуток = {a*b})")
```
