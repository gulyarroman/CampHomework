```python
import math

def dp(floors, balls):
    INF = float('inf')
    t = [[INF]*(floors+1) for _ in range(balls+1)]
    for f in range(floors+1): t[0][f] = 0 if f==0 else INF; t[1][f] = f
    for n in range(balls+1): t[n][0] = 0
    for n in range(2, balls+1):
        for f in range(2, floors+1):
            best, lo, hi = INF, 1, f
            while lo <= hi:
                mid = (lo+hi)//2
                w = max(t[n-1][mid-1], t[n][f-mid])
                if w < best: best = w
                if t[n-1][mid-1] < t[n][f-mid]: lo = mid+1
                else: hi = mid-1
            t[n][f] = 1+best
    return t

def checkpoints(t, floors, balls):
    pts, rem, off = [], floors, 0
    for _ in range(t[balls][floors]+2):
        if rem <= 0: break
        k = t[balls][rem]
        for x in range(1, rem+1):
            if 1+max(t[balls-1][x-1], t[balls][rem-x]) == k:
                pts.append(off+x); off += x; rem -= x; break
    return pts

def show(floors, balls):
    t = dp(floors, balls)
    drops = t[balls][floors]
    lb = math.ceil(math.log2(max(floors,2)))
    cp = checkpoints(t, floors, balls)
    gaps = [cp[0]] + [cp[i]-cp[i-1] for i in range(1,len(cp))]
    print(f'floors={floors}  balls={balls}  worst-case drops={drops}  log2-lower-bound={lb}')
    print(f'checkpoints : {cp}')
    print(f'gap sizes   : {gaps}')
    if balls==1: print('strategy    : linear scan')
    elif balls==2: k=drops; print(f'strategy    : triangular  k(k+1)/2>={floors}  k={k}')
    else: print(f'strategy    : DP-optimal')
    print()
```

```python
# preset cases
for b, f in [(2,10),(2,100),(3,100),(5,1000)]:
    show(f, b)
```

```python
# interactive input
while True:
    try:
        line = input('Enter floors balls (or q to quit): ').strip()
        if line.lower() in ('q','quit','exit'): break
        f, b = map(int, line.split())
        assert f>=1 and b>=1
        show(f, b)
    except (ValueError, AssertionError):
        print('invalid input — enter two positive integers, e.g.: 100 2')
    except EOFError:
        break
```
